```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen('https://dunststudio.com/product/list.html?cate_no=589')
bs = BeautifulSoup(html, 'html.parser')
```

```python
product_list = bs.find_all('li', {'class' : 'xans-record-'})
```

```python
detail_link = []
for product in product_list:
    product_link = product.find('a', class_='noborder')
    if product_link:
        href = product_link.get('href')
        detail_link.append(href)

detail_link
```
